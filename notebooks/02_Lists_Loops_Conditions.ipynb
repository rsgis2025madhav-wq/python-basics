## 6. Mini Task â€“ Rainfall classification
### ðŸ”¶ Task 6.1 â€“ Classify rainfall intensity

```python
rain = [0.0, 12.5, 34.0, 2.0, 0.0, 55.1, 8.4]

no_rain = 0
light = 0
moderate = 0
heavy = 0
threshold_light = 8.4
threshold_moderate = 34.0
threshold_heavy = 55.1

for r in rain:
    if r == 0.0:
        no_rain += 1
    elif r < threshold_light:
        light += 1
    elif r < threshold_moderate:
        moderate += 1
    elif r < threshold_heavy:
        heavy += 1
    else:
        heavy +=1

print("No rain days:", no_rain)
print("Light rain days:", light)
print("Moderate rain days:", moderate)
print("Heavy rain days:", heavy)
```
